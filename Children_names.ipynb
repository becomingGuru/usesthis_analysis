```python
years = range(1880,2012)
frames = []
for year in years:
    frame = pd.read_csv('/Users/lpd/Downloads/names/yob%d.txt'%year,names=['name','sex','births'])
    frame['year'] = year
    frames.append(frame)
```

```python
names = pd.concat(frames, ignore_index=True)
```
